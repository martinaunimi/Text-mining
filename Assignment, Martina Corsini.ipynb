```python
from sklearn.base import BaseEstimator
from sklearn.base import TransformerMixin
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
class CleanText( BaseEstimator, TransformerMixin ):
    
    def __init__( self, lang = "english"):
        self.lang = lang
    
    def fit( self, X, y = None ):
        return self 
   

    def clean( self, x ):
        words   = [word for word in word_tokenize(x.lower()) if word.isalpha() and word not in stopwords.words("english")]
        return " ".join(words)
    
    def transform( self, X, y = None ):
        return X["text"].apply(self.clean)
```

```python
class CustomFeatures( BaseEstimator, TransformerMixin ):
     
    def __init__( self ):
        return
      
    def fit( self, X, y = None ):
        return self 
        
    def transform( self, X, y = None ):
        f           = pd.DataFrame()
        f['len']    = X['text'].str.len()
        f['digits'] = X['text'].str.findall(r'\d').str.len()
        f['punct']  = X['text'].str.findall(r'[^a-zA-Z\d\s:]').str.len()
        return f[['len','digits','punct']]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.pipeline import FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import StandardScaler
pipe = Pipeline([("extract", FeatureUnion([("terms", Pipeline([('clean', CleanText()), 
                                                               ('tfidf', TfidfVectorizer())])),
                                           ("custom", CustomFeatures())])),
                 ("select", SelectKBest(score_func = chi2)),
                 ("scale", StandardScaler(with_mean = False))])
#text is clean (no more stop words and punctuaction), tokenized and vectorized (with TF-IDF method,useful for the classification)
```

Dataset: Movies Titles (https://www.kaggle.com/arch11/movies-titles?select=Movie_Id_Titles.csv)

Topic modeling is an unsupervised machine learning technique that analyzes text data to determine word clusters for a set of documents. The clusters of similar words are called topics. We choose the number of topics at the beginning of our process. Latent Dirichlet Allocation (LDA) is one of the most popular topic modeling methods.

```python
#clean and tokenize
import nltk
import spacy
from spacy.lang.en import English
parser = English()
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
#filter out stop words
en_stop = set(nltk.corpus.stopwords.words('english'))
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
#import dataset prepared for LDA
text_data = []
with open(r'C:\Users\DELL\Desktop\Titles.csv',encoding="utf8") as f:
    for line in f:
        tokens = prepare_text_for_lda(line)
        text_data.append(tokens)        
text_data
```
